## Media Geometrica

A media e calculada pela multiplicacao de todos os dados e em seguida retirada a raiz de N da mesma.



A Media Geometrica possui muita aplicacao na geometria, para por exemplo comparar lados de prismas, outra area de utilizacao e' na Matematica Financeira, que envolvem taxa percentual acumulada. (como o grafico abaixo)

### Vamos ver na pratica.

```python
# Utilizaremos de uma funcao pronta do scipy
from scipy.stats.mstats import gmean
```

```python
# Realizaremos os calculos no mesmo Dataset do Arquivo de Media_Moda_Mediana
import numpy as np
dados = np.array([150, 151, 152, 152, 153, 154, 155, 155, 155, 155, 156, 156, 156,
                  157, 158, 158, 160, 160, 160, 160, 160, 161, 161, 161, 161, 162,
                  163, 163, 164, 164, 164, 165, 166, 167, 168, 168, 169, 170, 172,
                  173])
```

```python
print('A Media Geometrica do DataSet acima resulta em :',gmean(dados))
```

## Media harmonica
Esta media, bastante utilizada por profissionais de Data Science principalmente para havaliar modelos de Machine Learning (para calculos de Precision e Recall), consiste em dividir o numero de elementos, pela somatoria dos dados, estando cada um deles elevado a menos um [  ^(-1)  ]. Como na Imagem a seguir:


```python
from scipy.stats.mstats import hmean
```

```python
print('A Media Harmonica do DataSet acima resulta em :',hmean(dados))
```

## Media Quadratica
Um dos calculos de media mais utilizado na fisica, e utilizada diariamente por profissionais de Data Science principalmente para havaliacao de Modelos de Regressao. Este e' calculado pela formula abaixo:


```python
# Criaremos uma funcao para calculo da Media Quadratica
import math
def qmean(dados):
    return math.sqrt(sum(n * n for n in dados) / len(dados))
```

```python
qmean(dados)
```

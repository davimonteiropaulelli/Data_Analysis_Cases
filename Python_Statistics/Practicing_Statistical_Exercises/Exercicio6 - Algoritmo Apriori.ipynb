# Resolucao

```python
df = pd.read_csv('census.csv')
df
```

```python
# O algoritimo so consegue entender dados no formato str ou obj. 
#Portanto sera necessario criar um DF com os dados a serem utilizados para teste.
def idade(column):
    if column < 18:
        return 'Jovem'
    elif column >= 18 and column <= 25:
        return 'Adulto I'
    elif column >= 26 and column <= 40:
        return 'Adulto II'
    elif column >= 41 and column <= 60:
        return 'Idoso I'
    elif column >= 61 and column <= 90:
        return 'Idoso II'
    else:
        return 'Nao informado'
    
df['age'] = df['age'].apply(lambda x : idade(x))
df.head(7)
```

```python
set(df['education-num'])
```

```python
def educacao(column):
    if column < 10:
        return 'Pouco Estudo'
    else:
        return 'Bem Estudado'
    
df['education-num'] = df['education-num'].apply(lambda x : educacao(x))
df.head(7)
```

```python
df = df[['age','workclass','education','education-num','marital-status','occupation','relationship','race','sex','native-country','income']]
df.head(7)
```

```python
# Separando uma menor quantidade para ser analisada (Neste caso, onde a maquina nao e' tao boa)
df = df.sample(2500)
```

```python
transacoes = []
for i in range(df.shape[0]):
    transacoes.append([str(df.values[i,j]) for j in range(df.shape[1])])
```

```python
transacoes
```

Uma vez o Novo DataFrame consolidado e preparado, o que resta e' submete-lo ao algoritmo.

```python
insight = apriori(transacoes,min_support=0.3,min_confidence=0.7)
resultado = list(insight)
print(len(resultado))
resultado
```

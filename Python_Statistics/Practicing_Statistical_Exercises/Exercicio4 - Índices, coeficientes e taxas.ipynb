## Importando bibliotecas

```python
import pandas as pd
import numpy as np
```

```python
# criando DataFrame
data = {'Ano graduacao':['1º','2º','3º','4º'],
       'Matriculas marco':[70,50,47,23],
       'Matriculas novembro':[65,48,40,22]}
df = pd.DataFrame(data = data,index=data['Ano graduacao'])
df = df.drop('Ano graduacao',axis=1)
df
```

```python
total = [df['Matriculas marco'].sum(),df['Matriculas novembro'].sum()]
total
```

```python
df['Taxa de Evasao'] = (round((df['Matriculas marco'] - df['Matriculas novembro']) / (df['Matriculas marco']) * 100,2))
```

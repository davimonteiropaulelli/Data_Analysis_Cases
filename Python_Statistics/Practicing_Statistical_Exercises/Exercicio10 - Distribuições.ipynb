Import

```python
import pandas as pd
import seaborn as sns
```

```python
# Carregando DataFrame
df = pd.read_csv('census.csv')
df.head()
```

```python
# Observar tipos de dados
df.dtypes
```

```python
# Histograma age
sns.distplot(df['age'])
```

Distribuicao GAMA

```python
# Histograma final-weight
sns.distplot(df['final-weight'])
```

Distribuicao GAMA

```python
# Histograma education-num
sns.distplot(df['education-num'],kde=False)
```

Distribuicao Binomial ou Distribuicao Poisson

```python
# Histograma capital-gain
sns.distplot(df['capital-gain'],kde = False)
```

Distribuicao Exponencial

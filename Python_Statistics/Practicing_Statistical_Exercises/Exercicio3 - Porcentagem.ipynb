# Importando Bibliotecas

```python
import pandas as pd
```

```python
df = pd.read_csv('census.csv')
df
```

```python
df2 = df[['education','income']]
df2
```

```python
df_g = df2.groupby(['education','income'])['education'].count()
df_g
```

```python
df_g.index
```

```python
df_g.max()
```

```python
df_g.min()
```

```python
# Comparando porcentagem de renda para pessoas que finalizam apenas o Ensino Medio (High School)
menor = round((df_g[' HS-grad', ' <=50K']) / (df_g[' HS-grad', ' <=50K'] + df_g[' HS-grad', ' >50K']) * 100,2)
maior = round((df_g[' HS-grad', ' >50K']) / (df_g[' HS-grad', ' <=50K'] + df_g[' HS-grad', ' >50K']) * 100,2)
print('Se voce apenas finalizou o High School(Ensino Medio) suas probabilidades sao:\n')
print('{}% de chances de ganhar menos ou igual a 50mil dolares por ano'.format(menor))
print('{}% de chances de ganhar mais de 50mil dolares por ano'.format(maior))
```
